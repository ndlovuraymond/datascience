```python
#splitting the data into train and test, split date is March 15 2016
split_date = '2016-03-15'
data_train = lbj_points.loc[lbj_points.index <= split_date].copy()
data_test = lbj_points.loc[lbj_points.index > split_date].copy()
```

```python
#function to create features for data
def create_features(df, label=None):
    df['date'] = df.index
    df["outcome"] = df["outcome"]
    df["Prev FT"] = df["Prev FT"]
    df["Prev FTA"] = df["Prev FTA"]
    df["Prev Three Att"] = df["Prev Three Att"]
    df["Prev Three"] = df["Prev Three"]
    df["Prev TOV"] = df["Prev TOV"]
    df["Prev FG"] = df["Prev FG"]
    df["Prev FGA"] = df["Prev FGA"]
    df["Prev STL"] = df["Prev STL"]
    df["Prev BLK"] = df["Prev BLK"]
    df["Prev Game Score"] = df["Prev Game Score"]
    df["Prev Plus Minus"] = df["Prev Plus Minus"]
    df["Prev ORB"] = df["Prev ORB"]
    df["Prev DRB"] = df["Prev DRB"]
    df["Prev AST"] = df["Prev AST"]
    df["Prev TRB"] = df["Prev TRB"]
    df["Prev MP"] = df["Prev MP"]
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    

    X = df[['outcome',"Prev FT","Prev FTA","Prev TRB","Prev ORB","Prev DRB","Prev TOV","Prev Plus Minus",
        "Prev FGA","Prev FG","Prev BLK","Prev STL","Prev MP","Prev Three","Prev Three Att","Prev Game Score",
            'hour','dayofweek','quarter','month','year','dayofyear','dayofmonth']]
    if label:
        y = df[label]
        return X, y
    return X
```

```python
#Applying the function to my data
X_train, y_train = create_features(data_train, label='pts')
X_test, y_test = create_features(data_test, label='pts')
X_train.head()
```

```python
#importing XGBoost model
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#instantiating model
reg = xgb.XGBRegressor(n_estimators=1000)
```

```python
#fitting data into the model
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
       verbose=False)
```

```python
import shap
```

```python
# train XGBoost model
X,y = shap.datasets.adult()
model = xgboost.XGBClassifier().fit(X, y)

# compute SHAP values
explainer = shap.Explainer(model, X)
shap_values = explainer(X)
```

```python
#checking feature important
_ = plot_importance(reg,height=0.9)
```

```python
#making prediction and plotting it against everything else
data_test['points_Prediction'] = reg.predict(X_test)
data_all = pd.concat([data_test, data_train], sort=False)
```
